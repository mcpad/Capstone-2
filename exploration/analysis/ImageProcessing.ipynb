```python
#create the image matrices for use in image classification model
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
directory = '../../../Capstone2Data/chictopia/'
```

```python
from keras.preprocessing.image import img_to_array
data = []
fileNames = []

for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith('.jpg') and len(data) < 10000:
            print(len(data))
            image = cv2.imread(os.path.join(root,file))
            image = cv2.resize(image, (100,150))
            image = img_to_array(image)
            data.append(image)
            fileNames.append(file)
        if len(data) > 10000:
            break
```

```python
#make x and y data
#load df
#query df - find index from fileNames
#reorder jpg array
#delete from df if not found
#y data tags only to start
```

```python
df = pd.read_csv('../data/data_training10000.csv')
df.head()
```

```python
#data for PCA
X = []
Y = []

for name in fileNames:
    idx = df.index[df['pictures'] == name]
    for i in idx:
        Y.append(df.loc[i]['tags'])
        X.append(data[fileNames.index(name)])
    
```

```python
#data for CNN
X2 = []
Y2 = []

for name in fileNames:
    idx = df.index[df['pictures'] == name]
    y = []
    for i in idx:
        y.append(df.loc[i]['tags'])
    X2.append(data[fileNames.index(name)])
    Y2.append(y)
```

```python
classes1500 = ['purse', 'hat', 'pink', 'blazer', 'white', 'tawny', 'gold', 't-shirt', 'green', 'belt', 'ivory', 'blouse', 'leggings', 'leather', 'tights', 'jumper', 'red', 'yellow', 'coat', 'dress', 'lace', 'shirt', 'scarf', 'vest', 'vintage', 'skirt', 'tan', 'cardigan', 'shoes', 'print', 'camel', 'sweater', 'pants', 'jeans', 'boots', 'accessories', 'sandals', 'gray', 'bag', 'denim', 'wedges', 'silver', 'brown', 'gum', 'navy', 'orange', 'beige', 'necklace', 'cream', 'sunglasses', 'heels', 'aquamarine', 'purple', 'floral', 'blue', 'shorts', 'jacket', 'crimson', 'eggshell', 'black']
classes5000 = ['hat', 'pink', 'blazer', 'white', 't-shirt', 'green', 'blouse', 'leather', 'red', 'coat', 'dress', 'shirt', 'skirt', 'shoes', 'sweater', 'pants', 'jeans', 'boots', 'gray', 'bag', 'brown', 'navy', 'beige', 'sunglasses', 'heels', 'blue', 'shorts', 'jacket', 'black']
classes10000 = ['hat', 'pink', 'blazer', 'white', 'red', 'coat', 'dress', 'shirt', 'skirt', 'shoes', 'sweater', 'jeans', 'boots', 'gray', 'bag', 'brown', 'navy', 'blue', 'shorts', 'jacket', 'black']
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb2 = MultiLabelBinarizer(classes = classes10000)
mlb2.fit(Y2)
Y2 = mlb2.transform(Y2)
```
